```python
from PFDFile import *
import numpy as np
import os
import pylab as plt
```

```python
def make_blocks(image):
	l = []
	if image.shape[1] < 60:
		return tuple(np.repeat(0, 56))
	for x in range(8):
		for y in range(6):
			l.append(np.mean(image[x*4:x*4+3, y*10:y*10+9])) 
	return tuple(l) # 56 dim
```

```python
def gather_feature(dir):
	X = []
	for file in os.listdir(dir):
		cand = PFD(dir + './' + file)
		print(len(X), dir + './' + file)
		try:
			subint_dim = cand.get_subints().shape[1]
			subband = cand.get_subbands()
		except:
			subint_dim = cand.get_subints().shape[1]
			subband = cand.get_subbands()
		#band_blocks = make_blocks(subband)
		subint_avg = np.mean(cand.get_subints())
		subint_medium = np.median(cand.get_subints())
		pro_len = len(cand.getprofile())
		pro_avg = sum(cand.getprofile()) / pro_len
		chi2 = cand.calc_redchi2()
		a, b = cand.plot_chi2_vs_DM(0, 800)
		chi_var = np.var(a)
		diff = np.diff(a)
		diff_max = np.max(diff)
		diff_min = np.min(diff)
		diff_avg = np.sum(diff) / len(diff)
		diff_larger = sum(np.abs(diff) > 0.5)
		diff_var = np.var(diff)
		sample = np.array([subint_dim, subint_avg, subint_medium, pro_avg, pro_len, 
                           chi2, chi_var, diff_max, diff_min, diff_avg, diff_larger, diff_var])
		#np.append(sample, band_blocks)
		X.append(sample)
		#if len(X) == 10:
		#	break
	X = np.array(X)
	print(X.shape)
	return X
```

```python
def save_data(X, Y):
	from sklearn.externals import joblib
	with open('12feat.fzy', 'wb') as fo: 
		joblib.dump((X, Y), fo)
	print('data saved')
```

```python
Xn = gather_feature('./p309n_pfd')
Yn = np.repeat(0, Xn.shape[0])
Yn = Yn.reshape(Yn.shape[0], 1)
Xp = gather_feature('./p309p_pfd')
Yp = np.repeat(1, Xp.shape[0])
Yp = Yp.reshape(Yp.shape[0], 1)
X = np.vstack((Xn, Xp))
Y = np.vstack((Yn, Yp))
save_data(X, Y)
```
