```python
def make_data(dir, pular, most_file_num=100):
	X = 0
	cnt = 0
	for file in os.listdir(dir):
		if cnt >= most_file_num:
			break
		print(cnt, file)
		cnt += 1

		cand = PFD(dir + './' + file)

		# np.array (32, ?) ? 
		#subband = cand.get_subbands()
		#subband_sum = np.sum(subband, axis=1)

		# np.array (64, ?) ? 
		#subints = cand.get_subints()
		#subints_sum = np.sum(subints, axis=1)

		# list of length ?
		#profile = cand.getprofile()
		#profile_sum = np.sum(profile)

		# a number ? 
		#chi2 = cand.calc_redchi2()

		# row and column axis number
		tmp1, tmp2 = cand.plot_chi2_vs_DM(0, 1500)

		x = np.array([tmp1])
		#x = np.hstack((x, subband_sum))

		if X is 0:
			X = x
		else:
			X = np.vstack((X, x))
	if pular is True:
		Y = np.repeat(1, X.shape[0])
	else:
		Y = np.repeat(0, X.shape[0]).reshape(X.shape[0], 1)
	return X, Y
```

```python
def save_data(X, Y):
	from sklearn.externals import joblib
	with open('100d-chi2.fzy', 'wb') as fo: 
		joblib.dump((X, Y), fo)
	print('data saved')
```
